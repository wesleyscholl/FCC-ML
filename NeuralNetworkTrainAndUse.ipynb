```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# Define the input shape using an Input object
input_tensor = keras.Input(shape=(1,))
```

```python
# Create your Sequential model, starting with the Input object
model = tf.keras.Sequential([
    input_tensor,  # Input layer defining the shape of the input data
    keras.layers.Dense(units=1)  # Dense layer with 1 unit (output)
])
```

```python
# Compile the model
# Optimizer: Stochastic Gradient Descent (sgd)
# Loss function: Mean Squared Error (mse)
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Create training data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)  # Input features
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)  # Target values
```

```python
# Train the model
# epochs: Number of training iterations over the entire dataset
model.fit(xs, ys, epochs=500)
```

```python
# Make a prediction
# Input: A NumPy array with a single value (10.0)
# To see the output, run the code.
print(model.predict(np.array([10.0])))
```
